# Normality Check

Now we can check the distribution of some of the columns to gain more information about our dataset.

## **Shapiro-Wilk test**

Assumption of this test: Observations in each sample are independent and distributed identically.

Hypothesis:

$h_0$: the sample has a Gaussian distribution.

$h_1$: the sample does not have a Gaussian distribution.


```python
from scipy.stats import shapiro
```

```python
#get a copy of our data frame again
norm_df = new_df.copy(deep=True)
```

Since this test is only applicable for numerical features , we can extract our numerical columns and apply this test on them.

```python
numerical_cols = norm_df.select_dtypes(include=['int64', 'float64']).columns
```

Now we apply the shpiro-wilk test.

```python
for i in numerical_cols:
    statistic, p_value = shapiro(norm_df[i])
    print(f'{i}: statistic={statistic:.2f}, p-value={p_value:.2f}')
```

## **Anderson-darling test**

This test also tells us if a specific column has a normal distribution or not.

```python
from scipy.stats import anderson
for i in numerical_cols :
  result = anderson(norm_df[i])
  print('Test statistic: %.3f' % result.statistic)
  print('Critical values: %s' % result.critical_values)
  print("*****")
```

## **Anova test**

We use this test for those columns that aren't numerical and therefore pearson test cannot be used for them.


```python
anova_df = new_df.copy(deep=True)
```

```python
from scipy import stats
cat = anova_df.groupby('Artist Name(s)')['Popularity']
f_value , p_value = stats.f_oneway(*[group for name , group in cat])
print('f={0:.3f} , p={1:.3f}'.format(f_value,p_value))
```

Results show us that there might be a realtion between these two columns.

# T-test

We use t-test to figure out if two groups of data are different from each other or not and if they are , how significant this difference is.

Hypothesis:

$h_0$ : There's no significant difference between two groups.

$h_1$ : There is a significant difference between them.


First we have to choose some genres to have a comparison of popularity between those specific genres.

```python
#let's get the list of all genres and then randomly choose some of them
genres_list = new_df['Artist Genres'].unique()
genres_list
```

```python
#now choose 5 random elements of genres
import random
new_list = list(genres_list)
new_list
```

```python
random_elements = random.sample(new_list, k=5)
```

```python
print(random_elements)
```

Now from this random selected genres I choose : disco , pop , hip hop , r&b , electropop

```python
from scipy.stats import ttest_ind
t_df = new_df[["Artist Genres", "Popularity"]]
t_df.head()
```

Now since genres column is a categorical feature in order to perform t-test on it we should groupy popularity by our chosen genres.

```python
chosen_genres =['pop','hip hop','dance pop','r&b','electropop']
grouped_data = [new_df[new_df["Artist Genres"].str.contains(genre)]["Popularity"] for genre in chosen_genres]
```

```python
t_value, p_value = ttest_ind(grouped_data[0], grouped_data[1])
print("T-value:", t_value)
print("P-value:", p_value)
```

As you can clearly see our p-values is less that 0.05 so the null hypothesis is rejected and there's definitely a difference between these two.
