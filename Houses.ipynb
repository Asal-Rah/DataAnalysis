## **Anderson-Darling test**

There exist various test to check if a data has a normal distribution or not.
Here we're using Anderson-Darling test on 'overallQual'.

```python
from scipy.stats import anderson
col_norm = df['OverallQual']
```

```python
result = anderson(col_norm)
```

```python
print('Test statistic: %.3f' % result.statistic)
print('Critical values: %s' % result.critical_values)
```

For better and clearer image of the feature's distribution we can plot it.

```python
import matplotlib.pyplot as plt
plt.hist(df['OverallQual'], bins=10 , color = 'purple' , edgecolor = 'black')
plt.title('Overall Quality')
plt.show()
```

Since the test statistic is much larger than any of the critical values, it means that the null hypothesis is rejected and this feature doesn't follow a normal distribution.

To get more information of our dataset we can check the normality of other colums as well.

```python
col_norm = df['BedroomAbvGr']
result = anderson(col_norm)
print('Test statistic: %.3f' % result.statistic)
print('Critical values: %s' % result.critical_values)
```

## **Shapiro-Wilk test**

We can also use other normality test such as Shapiro-Wilk test.

Assumption :
Observations in each sample are independent and distributed identically.

This test has two hypothesis:

$h_0$: the sample has a Gaussian distribution.

$h_1$: the sample does not have a Gaussian distribution.



```python
from scipy.stats import shapiro
col_norm = df['LotArea']
statistic, p_value = shapiro(col_norm)
print(f'Statistic: {statistic:.3f}')
print(f'p-value: {p_value:.3f}')
```

Since our p-value is less than 0.05 we can say that 'LotArea' is not normally distributed.

## Anova **test**

The previous test only works for numerical features and for categorical ones we must try other methods. Here we group our target column by the neighborhood's different values and then we apply **anova** test.

This test has two hypothesis:

$h_0$ : There is no significant difference between the means of the features being compared.

$h_a$ : There is a significant difference between them.

```python
cat_df = df.copy(deep=True)
```

```python
from scipy import stats
cat = cat_df.groupby('Neighborhood')['SalePrice']
f_value , p_value = stats.f_oneway(*[group for name , group in cat])
print('f={0:.3f} , p={1:.3f}'.format(f_value,p_value))
```

The shown results lead to the fact that there is a significant difference between the means of two features, so these two features are possibly correlated in some way.

```python
cat = cat_df.groupby('SaleCondition')['SalePrice']
f_value , p_value = stats.f_oneway(*[group for name , group in cat])
print('f={0:.3f} , p={1:.3f}'.format(f_value,p_value))
```

As you can see sale condition is also related to sale price but their relation might be less strong compared to Neighborhoof and Sale price bacause of lower f-value.

# T-test

We use t-test to figure out if two groups of data are different from each other or not and if they are , how significant this difference is.


```python
from scipy.stats import ttest_ind
t_df = df.copy(deep=True)
```

```python
t_statistic, p_value = ttest_ind(t_df['SalePrice'], t_df['2ndFlrSF'])
print('T-Statistic:', t_statistic)
print('P-Value:', p_value)
```

This result makes us certain that there's a large difference between the means of the two columns we chose . Since the p-value is 0.0 we can coclude that this difference hasn't happen by chance.
